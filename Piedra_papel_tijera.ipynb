# Piedra Papel & Tijera

- Reglas del Juego "Piedra, Papel y Tijera":

    - Dos jugadores eligen una de las tres opciones: "Piedra", "Papel" o "Tijera".

    - Las reglas para determinar al ganador son:
        
        - Piedra vence a Tijera.

        - Tijera vence a Papel.

        - Papel vence a Piedra.

        - El jugador que gana una ronda obtiene un punto.
        
        - El juego continÃºa hasta que un jugador haya acumulado 3 puntos, convirtiÃ©ndose en el ganador.


- Pistas para las alumnas:

    1. Comienza por definir el nÃºmero de puntos objetivo que se deben alcanzar para ganar el juego.

    2. Implementa una funciÃ³n que obtenga la elecciÃ³n del jugador y valide que sea una de las opciones vÃ¡lidas ("Piedra", "Papel" o "Tijera").

    3. Utiliza un bucle `while` para llevar un registro de los puntos de cada jugador y continuar el juego hasta que se alcance el nÃºmero de puntos objetivo.

    4. Genera una elecciÃ³n aleatoria para el segundo jugador (la computadora) utilizando el mÃ³dulo `random`.

    5. Determina al ganador de cada ronda y actualiza los puntos correspondientes.

    6. Verifica si algÃºn jugador ha alcanzado el nÃºmero de puntos objetivo y finalicen el juego cuando eso suceda.

```python
import random

# NÃºmero de puntos objetivo
puntos_objetivo = 3
puntos_jugador = 0 
puntos_pc = 0
ronda = 1

# Opciones vÃ¡lidas
opciones = ["piedra", "papel", "tijera"]

# Respuestas cuando el jugador gana
jugador_gana_ronda = ["BIEN HECHO!","LA RONDA ES TUYA!","ESTÃS DE SUERTE! HAS GANADO ESTA RONDA"]
jugador_pierde_ronda = ["OOPS...HAS PERDIDO", "TU OPONENTE HA GANADO ESTA PERO, PERO AÃšN TIENES CHANCES!", "HAS PERDIDO ESTA RONDA, ÃNIMO QUE EL JUEGO CONTINÃšA..."]

print( "Â¡BIENVENIDE A PIEDRA, PAPEL & TIJERA! \n\nEstas son las reglas del juego: \n    1) Dos jugadores eligen una de las tres opciones: Piedra, Papel o Tijera. \n    2) Las reglas para determinar al ganador son: \n       --> Piedra vence a Tijera. \n       --> Tijera vence a Papel. \n       --> Papel vence a Piedra. \n    3) El jugador que gana una ronda obtiene un punto. \n    4) El juego continÃºa hasta que un jugador haya acumulado 3 puntos, convirtiÃ©ndose en el ganador.\n\nÂ¡EMPIEZA EL JUEGO!ğŸ‘‡ ")
# ElecciÃ³n del jugador
while puntos_jugador < puntos_objetivo and puntos_pc < puntos_objetivo:

     print(f"\n-------------- RONDA #{ronda}---------------")
     print("\nElige piedrağŸª¨, papelğŸ§» o tijeraâœ‚ï¸...\n")
     eleccion_jugador = input("Elige piedra, papel o tijera: ").lower()

     if eleccion_jugador in opciones:
          ronda += 1
          print("Has elegido:", eleccion_jugador)
     else:
          print("OpciÃ³n no vÃ¡lida, prueba otra vez.")
          continue
          
     #Eleccion PC
     eleccion_pc = random.choice (opciones)
     print("Tu oponente ha elegido:", eleccion_pc)

     # Decidir ganador
     
     #empate
     if eleccion_jugador == eleccion_pc:
          print("Es un empate!ğŸ™ƒ")
          print(f"Tus puntos: {puntos_jugador} \nPuntos oponente: {puntos_pc}")
     #jugador gana
     elif eleccion_jugador == "piedra" and eleccion_pc == "tijera" or eleccion_jugador == "tijera" and eleccion_pc == "papel" or eleccion_jugador == "papel" and eleccion_pc == "piedra":
          print(random.choice(jugador_gana_ronda)) 
          puntos_jugador += 1
     
          print(f"Tus puntos: {puntos_jugador} \nPuntos oponente: {puntos_pc}")
     #jugador pierde
     else: 
          
          print(random.choice(jugador_pierde_ronda))
          puntos_pc += 1
          print(f"Tus puntos: {puntos_jugador} \nPuntos oponente: {puntos_pc}")

else:
     if puntos_jugador == puntos_objetivo:
          print("\n\nFELICIDADES! HAS GANADO EL JUEGO ğŸ†\n")
     else:
          print("\n\nHAS PERDIDO ğŸ˜­ MALA SUERTE...INTÃ‰NTALO DE NUEVO!\n")

```
